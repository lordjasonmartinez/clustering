```python
import pandas as pd

from sklearn.cluster import MiniBatchKMeans
```

```python
if __name__ == "__main__":
    dataset = pd.read_csv('./data/diamonds.csv')

    # Utiliza una lista para especificar las columnas que deseas eliminar
    X = dataset.drop(['cut', 'color','clarity'], axis=1)

    kmeans = MiniBatchKMeans(n_clusters=4, batch_size=8).fit(X)
    print("="*64)
    print("Total de centros: ", len(kmeans.cluster_centers_))
    print("="*64)
    print(kmeans.predict(X))
    print("="*64)

    dataset['group'] = kmeans.predict(X)
    print(dataset)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

if __name__ == "__main__":
    dataset = pd.read_csv('./data/diamonds.csv')

    # Utiliza una lista para especificar las columnas que deseas eliminar
    X = dataset.drop(['cut', 'color', 'clarity'], axis=1)

    kmeans = MiniBatchKMeans(n_clusters=4, batch_size=8).fit(X)
    
    # Aadir las predicciones al conjunto de datos
    dataset['group'] = kmeans.predict(X)

    # Configurar el estilo y los colores de los grficos
    sns.set(style="whitegrid")
    colors = ["black", "yellow", "blue", "orange"]

    # Scatter plot para visualizar las predicciones
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=dataset['group'], palette=colors, legend='full', alpha=0.8)
    plt.title('Resultados del K-Means')
    plt.xlabel(X.columns[0])
    plt.ylabel(X.columns[1])
    plt.show()
```

Mean-Shift

```python
from sklearn.cluster import MeanShift

if __name__ == "__main__":
    dataset = pd.read_csv('./data/diamonds.csv')

    # Utiliza una lista para especificar las columnas que deseas eliminar
    X = dataset.drop(['cut', 'color', 'clarity'], axis=1)

    meanshift = MeanShift().fit(X)  # Corregir aqu
    print("="*64)
    print(max(meanshift.labels_))
    print("="*64)
    print(meanshift.cluster_centers_)

    dataset['meanshift'] = meanshift.labels_
    print("="*64)
    print(dataset)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

if __name__ == "__main__":
    dataset = pd.read_csv('./data/diamonds.csv')

    # Utiliza una lista para especificar las columnas que deseas eliminar
    X = dataset.drop(['cut', 'color', 'clarity'], axis=1)

    meanshift = MeanShift().fit(X)
    
    # Aadir las etiquetas de los clusters a tu conjunto de datos
    dataset['meanshift'] = meanshift.labels_

    # Crear un grfico de dispersin con Seaborn
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=dataset, x='carat', y='price', hue='meanshift', palette=['yellow', 'black'], legend='full')

    # Ajustar la leyenda
    plt.legend(title='Clusters', loc='upper right')
    
    plt.title('Resultado de MeanShift')
    plt.xlabel('Carat')
    plt.ylabel('Price')
    plt.show()
```
